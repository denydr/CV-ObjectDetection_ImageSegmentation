#### PT.3 Deriving Bounding Boxes: Utilization of Binary Multi-Object Masks

```python
import cv2
import numpy as np
import json
from pathlib import Path

def derive_bounding_boxes(binary_mask, min_area=50):
    """
    Derives bounding boxes from a binary mask using connected components.
    Assumes that the binary mask has only two values: 0 (background) and 255 (object).
    
    Args:
        binary_mask (np.ndarray): A single-channel binary image.
        min_area (int): Minimum area (in pixels) to consider as a valid object.
        
    Returns:
        boxes (list): A list of bounding boxes, each represented as (x, y, w, h).
    """
    # Ensure the mask is of type uint8
    binary_mask = binary_mask.astype(np.uint8)
    
    # Run connected components analysis
    num_labels, labels, stats, _ = cv2.connectedComponentsWithStats(binary_mask, connectivity=8)
    
    boxes = []
    # Loop over each component (skip label 0, which is background)
    for i in range(1, num_labels):
        x, y, w, h, area = stats[i]
        if area >= min_area:
            boxes.append((int(x), int(y), int(w), int(h)))
    return boxes

def process_sequence_bounding_boxes(seq_folder, min_area=50):
    """
    Processes all binary mask images in a sequence folder to derive bounding boxes.
    
    Args:
        seq_folder (Path): Folder containing binary mask images for a sequence.
        min_area (int): Minimum area threshold.
        
    Returns:
        seq_bb (dict): Dictionary mapping each mask filename to its list of bounding boxes.
    """
    seq_bb = {}
    for mask_file in sorted(seq_folder.glob("*.*")):
        if mask_file.suffix.lower() not in ['.png', '.jpg', '.jpeg']:
            continue
        
        binary_mask = cv2.imread(str(mask_file), cv2.IMREAD_GRAYSCALE)
        if binary_mask is None:
            print(f"Warning: Could not load {mask_file.name}")
            continue
        
        boxes = derive_bounding_boxes(binary_mask, min_area=min_area)
        seq_bb[mask_file.name] = boxes
        print(f"Processed {mask_file.name}: found {len(boxes)} bounding boxes")
    return seq_bb

def process_all_sequences(input_root, output_json_path, min_area=50):
    """
    Processes all sequence folders under input_root, deriving bounding boxes for each binary mask,
    and aggregates the results in a dictionary.
    
    Args:
        input_root (Path): Root directory containing subfolders for each multi-object sequence.
        output_json_path (Path): Path to save the aggregated bounding boxes JSON.
        min_area (int): Minimum area threshold.
        
    Returns:
        all_bb (dict): Dictionary mapping sequence names to their bounding boxes data.
    """
    input_root = Path(input_root)
    all_bb = {}
    
    for seq_folder in sorted(input_root.iterdir()):
        if seq_folder.is_dir():
            print(f"Processing sequence: {seq_folder.name}")
            seq_bb = process_sequence_bounding_boxes(seq_folder, min_area=min_area)
            all_bb[seq_folder.name] = seq_bb
    
    # Optionally save to JSON
    output_json_path.parent.mkdir(parents=True, exist_ok=True)
    with open(output_json_path, "w") as f:
        json.dump(all_bb, f, indent=2)
    
    print("Saved bounding boxes JSON to:", output_json_path.resolve())
    return all_bb

# Example usage:
if __name__ == "__main__":
    # Root directory with binary masks for multi-object sequences
    input_root = Path("output") / "binary_multi_object_masks"
    # Where to save the aggregated bounding boxes JSON
    output_json_path = Path("output") / "bounding_boxes-multi_object.json"
    
    all_bounding_boxes = process_all_sequences(input_root, output_json_path, min_area=50)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

def derive_bounding_box(binary_mask, min_area=50):
    """
    Derives a bounding box from a binary mask using connected components.
    Returns the largest bounding box in the mask if multiple regions exist.
    Assumes the binary mask has values 0 and 255.
    """
    binary_mask = binary_mask.astype(np.uint8)
    num_labels, labels, stats, _ = cv2.connectedComponentsWithStats(binary_mask, connectivity=8)
    
    # Skip the background (label 0) and select the largest component (by area)
    best_box = None
    best_area = 0
    for i in range(1, num_labels):
        x, y, w, h, area = stats[i]
        if area >= min_area and area > best_area:
            best_area = area
            best_box = (int(x), int(y), int(w), int(h))
    return best_box, best_area

def overlay_boxes_with_labels(raw_frame_path, binary_mask_folder, base_filename, expected_labels, min_area=50):
    """
    Overlays bounding boxes with labels on the raw frame.
    
    Args:
        raw_frame_path (Path): Path to the raw frame image.
        binary_mask_folder (Path): Folder containing binary mask images for the frame.
        base_filename (str): Base filename (e.g., "00000") to match mask files like "00000_obj0.png".
        expected_labels (list): List of object names to assign to each bounding box.
        min_area (int): Minimum area threshold.
        
    Returns:
        frame (np.ndarray): The raw frame with bounding boxes and labels drawn.
        assigned (list): List of tuples (box, label).
    """
    # Find all binary mask files matching the pattern
    mask_files = sorted(list(binary_mask_folder.glob(f"{base_filename}_obj*.png")))
    boxes_info = []
    
    for mask_file in mask_files:
        binary_mask = cv2.imread(str(mask_file), cv2.IMREAD_GRAYSCALE)
        if binary_mask is None:
            print(f"Warning: Could not load {mask_file.name}")
            continue
        
        # Derive bounding box for this binary mask
        box, area = derive_bounding_box(binary_mask, min_area=min_area)
        if box is not None:
            boxes_info.append((mask_file.name, box, area))
    
    # If no boxes are found, return the raw frame unchanged
    if not boxes_info:
        print("No bounding boxes derived.")
        return cv2.imread(str(raw_frame_path)), []
    
    # Sort boxes by area in descending order (largest first)
    boxes_info.sort(key=lambda x: x[2], reverse=True)
    
    # Assign labels to boxes by order. If there are more boxes than labels, remaining get "unknown".
    assigned = []
    for idx, (fname, box, area) in enumerate(boxes_info):
        label = expected_labels[idx] if idx < len(expected_labels) else "unknown"
        assigned.append((box, label))
    
    # Load the raw frame.
    frame = cv2.imread(str(raw_frame_path))
    if frame is None:
        raise FileNotFoundError(f"Could not load raw frame from {raw_frame_path}")
    
    # Draw each bounding box with its label on the frame.
    for (x, y, w, h), label in assigned:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.putText(frame, label, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)
    
    return frame, assigned

# Example usage:
if __name__ == "__main__":
    # Path to the raw frame for the sequence (adjust as needed)
    raw_frame_path = Path("datasets/DAVIS_dataset/JPEGImages/480p/bike-packing/00000.jpg")
    
    # Folder where binary masks for multi-object sequences are saved (adjust path as needed)
    binary_mask_folder = Path("output/binary_multi_object_masks/bike-packing")
    
    # Base filename (e.g., "00000") for the frame's binary mask files
    base_filename = "00000"
    
    # Expected labels from your semantic JSON for the sequence "bike-packing" (example)
    expected_labels = ["bicycle", "person"]  # Adjust based on your semantics
    
    frame_with_labels, assigned_boxes = overlay_boxes_with_labels(raw_frame_path, binary_mask_folder, base_filename, expected_labels, min_area=50)
    
    # Display the result using matplotlib
    frame_rgb = cv2.cvtColor(frame_with_labels, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(8,6))
    plt.imshow(frame_rgb)
    plt.title("Raw Frame with Labeled Bounding Boxes")
    plt.axis("off")
    plt.show()
    
    # Print assigned bounding boxes and labels
    for box, label in assigned_boxes:
        print("Box:", box, "-> Label:", label)
```

##### Trying out how the boxes are generated on video sequences

```python
import cv2
import numpy as np
from pathlib import Path

def derive_bounding_boxes(binary_mask, min_area=50):
    """
    Derives bounding boxes from a binary mask using connected components.
    Assumes the binary mask contains only 0 (background) and 255 (object) values.
    
    Args:
        binary_mask (np.ndarray): A single-channel binary image.
        min_area (int): Minimum area in pixels to consider as an object.
        
    Returns:
        boxes (list): A list of bounding boxes, each represented as (x, y, w, h).
    """
    binary_mask = binary_mask.astype(np.uint8)
    num_labels, labels, stats, _ = cv2.connectedComponentsWithStats(binary_mask, connectivity=8)
    boxes = []
    for i in range(1, num_labels):  # skip background (label 0)
        x, y, w, h, area = stats[i]
        if area >= min_area:
            boxes.append((int(x), int(y), int(w), int(h)))
    return boxes

def overlay_boxes_on_frame(frame, boxes):
    """
    Draws bounding boxes on the frame.
    
    Args:
        frame (np.ndarray): Raw frame image.
        boxes (list): List of bounding boxes, each as (x, y, w, h).
        
    Returns:
        frame (np.ndarray): The frame with bounding boxes drawn.
    """
    for (x, y, w, h) in boxes:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
    return frame

def find_corresponding_mask(frame_file, mask_folder):
    """
    Finds a binary mask file in mask_folder that matches the base filename of frame_file,
    regardless of extension.
    
    Args:
        frame_file (Path): Path to the raw frame file.
        mask_folder (Path): Folder containing binary mask files.
        
    Returns:
        mask_path (Path or None): The path to the matching mask file, or None if not found.
    """
    stem = frame_file.stem
    # Look for any file with this stem in the mask folder
    candidates = list(mask_folder.glob(f"{stem}.*"))
    if candidates:
        return candidates[0]  # Return the first matching file
    return None

def process_video_sequence(raw_frames_folder, binary_masks_folder, output_video_path, min_area=50, fps=25):
    """
    Processes an entire video sequence by deriving bounding boxes from binary masks,
    overlaying them on raw frames, and writing the annotated frames to an output video.
    
    Args:
        raw_frames_folder (Path): Folder containing raw frame images.
        binary_masks_folder (Path): Folder containing corresponding binary mask images.
        output_video_path (Path): Path to save the annotated output video.
        min_area (int): Minimum area threshold for objects.
        fps (int): Frames per second for the output video.
    """
    raw_frames_folder = Path(raw_frames_folder)
    binary_masks_folder = Path(binary_masks_folder)
    
    # Get sorted list of frame files (we use any image format)
    frame_files = sorted(list(raw_frames_folder.glob("*.*")))
    if not frame_files:
        raise FileNotFoundError(f"No frames found in {raw_frames_folder}")
    
    # Initialize video writer using the size from the first frame
    first_frame = cv2.imread(str(frame_files[0]))
    if first_frame is None:
        raise FileNotFoundError(f"Could not load the first frame from {frame_files[0]}")
    height, width, _ = first_frame.shape
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out_video = cv2.VideoWriter(str(output_video_path), fourcc, fps, (width, height))
    
    for frame_file in frame_files:
        # Load the raw frame
        frame = cv2.imread(str(frame_file))
        if frame is None:
            print(f"Warning: Could not load raw frame {frame_file.name}")
            continue
        
        # Find the corresponding binary mask using the stem
        mask_path = find_corresponding_mask(frame_file, binary_masks_folder)
        if mask_path is None:
            print(f"Warning: Could not find binary mask for {frame_file.name}")
            boxes = []
        else:
            binary_mask = cv2.imread(str(mask_path), cv2.IMREAD_GRAYSCALE)
            if binary_mask is None:
                print(f"Warning: Could not load binary mask for {frame_file.name}")
                boxes = []
            else:
                boxes = derive_bounding_boxes(binary_mask, min_area=min_area)
        
        # Overlay bounding boxes on the raw frame
        annotated_frame = overlay_boxes_on_frame(frame.copy(), boxes)
        out_video.write(annotated_frame)
        
        # Optional: display frame (comment out in production)
        cv2.imshow("Annotated Frame", annotated_frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    
    out_video.release()
    cv2.destroyAllWindows()
    print(f"Annotated video saved to: {output_video_path}")

# Example usage:
if __name__ == "__main__":
    # For multi-object sequences, update the paths accordingly.
    # For example, for the "bike-packing" sequence:
    raw_frames_folder = Path("datasets/DAVIS_dataset/JPEGImages/480p") / "parkour"
    # Adjust the binary masks folder path (for example, binary masks are saved as PNG even if raw frames are JPG)
    binary_masks_folder = Path("output") / "binary_single_object_masks" / "parkour"
    output_video_path = Path("output") / "annotated_video.mp4"
    
    process_video_sequence(raw_frames_folder, binary_masks_folder, output_video_path, min_area=50, fps=25)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

def derive_bounding_boxes(binary_mask, min_area=50):
    """
    Derives bounding boxes from a binary mask using connected components.
    Assumes the binary mask contains only 0 (background) and 255 (object) values.
    
    Args:
        binary_mask (np.ndarray): A single-channel binary image.
        min_area (int): Minimum area (in pixels) to consider as an object.
        
    Returns:
        boxes (list): A list of bounding boxes, each represented as (x, y, w, h).
    """
    binary_mask = binary_mask.astype(np.uint8)
    num_labels, labels, stats, _ = cv2.connectedComponentsWithStats(binary_mask, connectivity=8)
    boxes = []
    for i in range(1, num_labels):  # skip label 0 (background)
        x, y, w, h, area = stats[i]
        if area >= min_area:
            boxes.append((int(x), int(y), int(w), int(h)))
    return boxes

def find_all_corresponding_masks(frame_file, mask_folder):
    """
    Finds all binary mask files in mask_folder that match the base filename of frame_file.
    For multi-object sequences, these files might be named "00000_obj0.png", "00000_obj1.png", etc.
    
    Args:
        frame_file (Path): Path to the raw frame file.
        mask_folder (Path): Folder containing binary mask files.
        
    Returns:
        mask_files (list): List of matching mask file paths.
    """
    stem = frame_file.stem
    # Look for files that start with the stem and have _obj in their name
    mask_files = sorted(list(mask_folder.glob(f"{stem}_obj*.png")))
    return mask_files

def overlay_boxes_on_frame(frame, boxes):
    """
    Draws bounding boxes on a frame.
    
    Args:
        frame (np.ndarray): The raw frame image.
        boxes (list): List of bounding boxes (each as (x, y, w, h)).
        
    Returns:
        frame (np.ndarray): The frame with drawn bounding boxes.
    """
    for (x, y, w, h) in boxes:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
    return frame

def process_video_sequence_multi_object(raw_frames_folder, binary_masks_folder, output_video_path, min_area=50, fps=25):
    """
    Processes an entire video sequence for multi-object masks:
      - For each raw frame, finds all corresponding binary mask files (for different objects).
      - Derives bounding boxes from each binary mask.
      - Overlays all bounding boxes on the raw frame.
      - Writes the annotated frame to an output video.
    
    Args:
        raw_frames_folder (Path): Folder containing raw frame images.
        binary_masks_folder (Path): Folder containing binary mask files for the sequence.
        output_video_path (Path): Path to save the annotated output video.
        min_area (int): Minimum area threshold for bounding boxes.
        fps (int): Frames per second for the output video.
    """
    raw_frames_folder = Path(raw_frames_folder)
    binary_masks_folder = Path(binary_masks_folder)
    
    frame_files = sorted(list(raw_frames_folder.glob("*.*")))
    if not frame_files:
        raise FileNotFoundError(f"No raw frames found in {raw_frames_folder}")
    
    # Initialize VideoWriter using the first frame's dimensions.
    first_frame = cv2.imread(str(frame_files[0]))
    if first_frame is None:
        raise FileNotFoundError(f"Could not load the first frame from {frame_files[0]}")
    height, width, _ = first_frame.shape
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out_video = cv2.VideoWriter(str(output_video_path), fourcc, fps, (width, height))
    
    for frame_file in frame_files:
        frame = cv2.imread(str(frame_file))
        if frame is None:
            print(f"Warning: Could not load raw frame {frame_file.name}")
            continue
        
        # Find all corresponding binary mask files for this frame
        mask_files = find_all_corresponding_masks(frame_file, binary_masks_folder)
        all_boxes = []
        if not mask_files:
            print(f"Warning: No binary masks found for {frame_file.name}")
        else:
            for mask_file in mask_files:
                binary_mask = cv2.imread(str(mask_file), cv2.IMREAD_GRAYSCALE)
                if binary_mask is None:
                    print(f"Warning: Could not load binary mask {mask_file.name}")
                    continue
                boxes = derive_bounding_boxes(binary_mask, min_area=min_area)
                all_boxes.extend(boxes)
        
        # Overlay all derived bounding boxes on the raw frame
        annotated_frame = overlay_boxes_on_frame(frame.copy(), all_boxes)
        out_video.write(annotated_frame)
        
        # Optional: display the frame for debugging
        cv2.imshow("Annotated Frame", annotated_frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    
    out_video.release()
    cv2.destroyAllWindows()
    print(f"Annotated video saved to: {output_video_path}")

# Example usage:
if __name__ == "__main__":
    # For multi-object sequences, update the folder paths accordingly.
    # Raw frames folder for the "bike-packing" sequence:
    raw_frames_folder = Path("datasets/DAVIS_dataset/JPEGImages/480p") / "bike-packing"
    # Binary masks folder for the "bike-packing" sequence:
    binary_masks_folder = Path("output") / "binary_multi_object_masks" / "bike-packing"
    # Output video path:
    output_video_path = Path("output") / "annotated_video_bike-packing.mp4"
    
    process_video_sequence_multi_object(raw_frames_folder, binary_masks_folder, output_video_path, min_area=50, fps=25)
```
