```python
import cv2
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt

def derive_bounding_boxes(binary_mask):
    """
    Derives bounding boxes from a binary mask using connected components.
    Assumes that 0 is background and non-zero (255) is the object.
    Returns a list of bounding boxes (x, y, w, h).
    """
    # Ensure binary_mask is of type uint8
    binary_mask = binary_mask.astype(np.uint8)
    
    # Use connected components to get all regions
    num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(binary_mask, connectivity=8)
    
    bounding_boxes = []
    # Skip the background (label 0)
    for i in range(1, num_labels):
        x, y, w, h, area = stats[i]
        # Optionally, filter out very small areas if needed
        if area > 50:
            bounding_boxes.append((x, y, w, h))
    return bounding_boxes

def process_and_visualize_bounding_boxes(input_dir):
    """
    Processes all binary masks in the given input directory, derives bounding boxes,
    and visualizes them on the corresponding raw frame if available.
    """
    input_dir = Path(input_dir)
    
    # Iterate over each sequence folder (each representing a video sequence)
    for sequence_folder in input_dir.iterdir():
        if sequence_folder.is_dir():
            print(f"Processing sequence: {sequence_folder.name}")
            # For demonstration, process the first image mask in the sequence folder.
            mask_files = sorted([f for f in sequence_folder.iterdir() if f.suffix.lower() in ['.png', '.jpg', '.jpeg']])
            if mask_files:
                sample_mask_path = mask_files[0]
                binary_mask = cv2.imread(str(sample_mask_path), cv2.IMREAD_GRAYSCALE)
                boxes = derive_bounding_boxes(binary_mask)
                print(f"Bounding boxes for {sample_mask_path.name}: {boxes}")
                
                # Optionally visualize the bounding boxes on the mask itself
                mask_with_boxes = cv2.cvtColor(binary_mask, cv2.COLOR_GRAY2BGR)
                for (x, y, w, h) in boxes:
                    cv2.rectangle(mask_with_boxes, (x, y), (x+w, y+h), (0, 255, 0), 2)
                
                # Display the image with bounding boxes using matplotlib
                plt.figure(figsize=(6, 6))
                plt.imshow(cv2.cvtColor(mask_with_boxes, cv2.COLOR_BGR2RGB))
                plt.title(f"Bounding Boxes in {sequence_folder.name}")
                plt.axis('off')
                plt.show()
            else:
                print(f"No mask files found in {sequence_folder}")

if __name__ == "__main__":
    # Input directory with binary masks (from your previous processing)
    binary_masks_dir = Path("output") / "binary_single_object_masks"
    process_and_visualize_bounding_boxes(binary_masks_dir)
```

```python
import cv2
import numpy as np
import json
from pathlib import Path

def derive_bounding_boxes(binary_mask, min_area=50):
    """
    Derives bounding boxes from a binary mask using connected components.
    Assumes that 0 is background and non-zero (255) is the object.
    Returns a list of bounding boxes, each as (x, y, w, h).
    """
    # Ensure the binary mask is of type uint8.
    binary_mask = binary_mask.astype(np.uint8)
    
    # Use connected components to get all regions in the binary mask.
    num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(binary_mask, connectivity=8)
    
    boxes = []
    # Skip the background label (0).
    for i in range(1, num_labels):
        x, y, w, h, area = stats[i]
        if area > min_area:  # filter out small regions if needed
            boxes.append((int(x), int(y), int(w), int(h)))
    return boxes

def save_bounding_boxes_for_sequence(seq_input_folder, seq_output_folder):
    """
    Processes all binary mask images in a sequence folder,
    derives bounding boxes for each image, and saves them in a JSON file.
    """
    # Ensure the output folder for this sequence exists.
    seq_output_folder.mkdir(parents=True, exist_ok=True)
    
    bb_data = {}  # dictionary to hold bounding boxes for each mask file
    
    # Process each mask image in the sequence folder.
    for mask_file in sorted(seq_input_folder.glob("*.*")):
        if mask_file.suffix.lower() in ['.png', '.jpg', '.jpeg']:
            mask = cv2.imread(str(mask_file), cv2.IMREAD_GRAYSCALE)
            if mask is None:
                print(f"Could not load {mask_file}")
                continue
            boxes = derive_bounding_boxes(mask)
            bb_data[mask_file.name] = boxes
    
    # Save the bounding boxes dictionary as a JSON file in the sequence output folder.
    output_json_path = seq_output_folder / "bounding_boxes.json"
    with open(output_json_path, "w") as f:
        json.dump(bb_data, f, indent=2)
    print(f"Saved bounding boxes for sequence '{seq_input_folder.name}' to {output_json_path}")

def process_all_sequences(input_root, output_root):
    """
    Iterates over all sequence folders in the input_root directory,
    derives bounding boxes for each binary mask image, and saves the results.
    """
    input_root = Path(input_root)
    output_root = Path(output_root)
    
    for seq_folder in input_root.iterdir():
        if seq_folder.is_dir():
            seq_output_folder = output_root / seq_folder.name
            print(f"Processing sequence: {seq_folder.name}")
            save_bounding_boxes_for_sequence(seq_folder, seq_output_folder)

if __name__ == "__main__":
    # Input directory with your binary masks for single-object sequences.
    binary_masks_root = Path("output") / "binary_single_object_masks"
    # Output directory for the bounding boxes (adjusted as requested).
    bounding_boxes_output_root = Path("output") / "bounding_boxes-single_object"
    
    process_all_sequences(binary_masks_root, bounding_boxes_output_root)
```

```python
import cv2
import numpy as np
import json
from pathlib import Path

def derive_bounding_boxes(binary_mask, min_area=50):
    """
    Derives bounding boxes from a binary mask using connected components.
    Assumes that 0 is background and non-zero (e.g., 255) is the object.
    Returns a list of bounding boxes, each as (x, y, w, h).
    """
    binary_mask = binary_mask.astype(np.uint8)
    num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(binary_mask, connectivity=8)
    boxes = []
    # Skip the background label (i=0)
    for i in range(1, num_labels):
        x, y, w, h, area = stats[i]
        if area > min_area:
            boxes.append((int(x), int(y), int(w), int(h)))
    return boxes

def get_bounding_boxes_for_sequence(seq_input_folder):
    """
    Processes all binary mask images in a sequence folder,
    derives bounding boxes for each image, and returns a dictionary
    mapping frame file names to a dictionary of bounding boxes keyed by an object ID.
    """
    bb_data = {}
    for mask_file in sorted(seq_input_folder.glob("*.*")):
        if mask_file.suffix.lower() in ['.png', '.jpg', '.jpeg']:
            mask = cv2.imread(str(mask_file), cv2.IMREAD_GRAYSCALE)
            if mask is None:
                print(f"Could not load {mask_file}")
                continue
            boxes_list = derive_bounding_boxes(mask)
            # Convert list of boxes to a dictionary using string keys "1", "2", ...
            boxes_dict = {str(i+1): box for i, box in enumerate(boxes_list)}
            bb_data[mask_file.name] = boxes_dict
    return bb_data

def process_all_sequences(input_root):
    """
    Iterates over all sequence subdirectories in the input_root,
    computes bounding boxes for each frame, and returns a single dictionary
    where keys are sequence names and values are the per-frame bounding boxes.
    """
    input_root = Path(input_root)
    all_bb = {}
    for seq_folder in input_root.iterdir():
        if seq_folder.is_dir():
            print(f"Processing sequence: {seq_folder.name}")
            seq_bb = get_bounding_boxes_for_sequence(seq_folder)
            all_bb[seq_folder.name] = seq_bb
    return all_bb

if __name__ == "__main__":
    # Input directory containing subdirectories of binary masks for single-object sequences.
    binary_masks_root = Path("output") / "binary_single_object_masks"
    # Process all sequences and get a global dictionary
    all_bounding_boxes = process_all_sequences(binary_masks_root)
    
    # Save the combined JSON in one file (structure similar to the multi-object JSON)
    output_path = Path("output") / "bounding_boxes_single_object.json"
    with open(output_path, "w") as f:
        json.dump(all_bounding_boxes, f, indent=2)
    
    print("Saved bounding boxes for single-object sequences to:", output_path.resolve())
```

#### ----NEW APPROACH---- PT 2. Deriving Bounding Boxes: Utilization of Label-to-Colour Images for Multi-Object Sequences

```python
import cv2
import numpy as np
import json
from pathlib import Path

# -------------------------------------
# Part 1: Compute Bounding Boxes from a Single Label Image
# -------------------------------------
def compute_bounding_boxes_from_label(label_img, min_area=100):
    """
    Computes bounding boxes from a label image.

    Args:
        label_img (np.ndarray): Single-channel label image where each pixel's value is an object label.
        min_area (int): Minimum area threshold to consider.

    Returns:
        boxes (dict): Mapping from label ID to bounding box (x, y, w, h).
    """
    boxes = {}
    unique_labels = np.unique(label_img)
    for label in unique_labels:
        if label == 0:
            continue  # Skip background
        ys, xs = np.where(label_img == label)
        if xs.size == 0 or ys.size == 0:
            continue
        x_min, x_max = int(xs.min()), int(xs.max())
        y_min, y_max = int(ys.min()), int(ys.max())
        w, h = x_max - x_min + 1, y_max - y_min + 1
        if w * h >= min_area:
            boxes[int(label)] = (x_min, y_min, w, h)
    return boxes

# -------------------------------------
# Part 2: Process All Label Images in a Sequence Folder
# -------------------------------------
def process_sequence_bounding_boxes(label_seq_folder, min_area=100):
    """
    Processes all label images in a sequence folder to compute bounding boxes.

    Args:
        label_seq_folder (str or Path): Folder containing label PNGs (e.g., *_label.png).
        min_area (int): Minimum area threshold.

    Returns:
        results (dict): { "00000.png": { label: [x, y, w, h], ... }, ... }
    """
    label_seq_folder = Path(label_seq_folder)
    results = {}

    # Find all *_label.png files
    label_files = sorted(label_seq_folder.glob("*_label.png"))

    for label_file in label_files:
        label_img = cv2.imread(str(label_file), cv2.IMREAD_UNCHANGED)
        if label_img is None:
            print(f"⚠️ Warning: Could not load label image {label_file.name}")
            continue

        # ✅ Remove "_label" so final JSON keys match real filenames like 00000.png
        clean_name = label_file.name.replace("_label", "")
        boxes = compute_bounding_boxes_from_label(label_img, min_area=min_area)
        results[clean_name] = boxes
        print(f"✅ Processed {label_file.name} → {clean_name}: {len(boxes)} boxes")
    
    return results

# -------------------------------------
# Part 3: Process All Sequences and Save to JSON
# -------------------------------------
def process_all_sequences_bounding_boxes(label_root_folder, min_area=100):
    """
    Processes all sequence subfolders and aggregates bounding boxes.

    Args:
        label_root_folder (str or Path): Root folder containing subfolders with label images.
        min_area (int): Minimum area threshold.

    Returns:
        all_results (dict): {
            "sequence_name": {
                "00000.png": { label: [x,y,w,h], ... }, ...
            },
            ...
        }
    """
    label_root_folder = Path(label_root_folder)
    all_results = {}

    for seq_folder in sorted(label_root_folder.iterdir()):
        if not seq_folder.is_dir():
            continue
        seq_name = seq_folder.name
        print(f"\n📁 Processing sequence: {seq_name}")
        seq_results = process_sequence_bounding_boxes(seq_folder, min_area=min_area)
        all_results[seq_name] = seq_results

    return all_results

# -------------------------------------
# Run the Pipeline
# -------------------------------------
if __name__ == "__main__":
    label_root_folder = Path("output/label_images/multi_object_annotations")  # 🔁 Your mask directory
    output_json_path = Path("output/bounding_boxes_from_label_images.json")
    min_area = 100  # Adjust if needed

    all_bb_results = process_all_sequences_bounding_boxes(label_root_folder, min_area=min_area)

    output_json_path.parent.mkdir(parents=True, exist_ok=True)
    with open(output_json_path, "w") as f:
        json.dump(all_bb_results, f, indent=2)

    print(f"\n✅ Saved bounding boxes JSON to: {output_json_path.resolve()}")
```

#### Deriving the JSON, including Bounding Boxes + Labels for Multi-Objecg Sequences

```python
import json
from pathlib import Path

# Paths to your input JSON files.
bb_json_path = Path("output/bounding_boxes_from_label_images.json")
semantics_json_path = Path("datasets/DAVIS_dataset/davis_semantics.json")

# Load the bounding boxes JSON
with open(bb_json_path, "r") as f:
    bb_data = json.load(f)

# Load the DAVIS semantics JSON
with open(semantics_json_path, "r") as f:
    semantics_data = json.load(f)

# Create a new dictionary that will include both bounding boxes and labels,
# preserving the per-object structure.
extended_bb = {}

# Process each sequence
for seq_name, frames in bb_data.items():
    extended_bb[seq_name] = {}
    
    # Get the DAVIS semantics mapping for the current sequence.
    # (The DAVIS JSON maps keys like "1" to "bicycle", "2" to "person", etc.)
    # We'll keep them as str -> str so that we can look up easily.
    if seq_name in semantics_data:
        sem_map = semantics_data[seq_name]
    else:
        sem_map = {}
    
    # Process each frame within the sequence
    for frame_name, boxes in frames.items():
        new_frame_boxes = {}
        for comp_key, bbox in boxes.items():
            # Look up the semantic label using the numeric key (if it exists).
            # comp_key is something like "1", "2", ...
            # We need to see if that comp_key is in sem_map.
            # If sem_map has int keys, convert comp_key -> int. Otherwise, check string.
            try:
                comp_key_int = int(comp_key)  # e.g., 1
            except ValueError:
                # If for some reason it's not a clean integer, fallback
                comp_key_int = None
            
            if comp_key_int is not None and str(comp_key_int) in sem_map:
                label = sem_map[str(comp_key_int)]
            else:
                # If we can't find it, mark it unknown
                label = f"unknown_{comp_key}"
            
            # Combine the label + numeric ID so we preserve uniqueness
            # E.g., "person_1", "person_2"
            final_key = f"{label}_{comp_key}"
            
            # Assign the bounding box under this new key
            new_frame_boxes[final_key] = bbox
        
        extended_bb[seq_name][frame_name] = new_frame_boxes

# Save the new extended JSON
output_path = Path("output") / "multi_object_boundingboxes_labels.json"
with open(output_path, "w") as f:
    json.dump(extended_bb, f, indent=2)

print("Saved extended bounding boxes with labels (preserving structure) to:", output_path.resolve())
```
